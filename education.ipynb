```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager
import plotly.graph_objects as go



df = {"SSC": 90, "INTERMEDIATE": 89.6, "B.SC(CS)" :70}
print(df)
```

```python
keys=list(df.keys())
values=list(df.values())
f = plt.figure()
f.set_figwidth(7)
f.set_figheight(5)
ax = plt.axes()
graph = plt.barh(keys,values,color=['green','blue','yellow'])
ax.set_facecolor("skyblue")
f.set_facecolor("skyblue")
plt.xlabel("Education",fontsize='18')
plt.ylabel("Percentage",fontsize='18')

i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()

    plt.text(x+width/2,
             y+height*1.01,
             str(values[i])+'%',
             ha='center',
             weight='bold',
             fontname='serif'
             )
    i += 1
plt.show()
plt.savefig('sample.jpg')
```
